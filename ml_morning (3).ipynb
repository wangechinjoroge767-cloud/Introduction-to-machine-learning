#### 1.1.1 Linear Regression.   
Linear regression is a supervised machine learning algorithm and statistical method used to model the relationship between a dependent(target) variable and one or more independent(predictor) variables.     

Linear regression is like trying to draw a line of best fit through a scatter of dots on a graph.  

_Formula_  
$y = mx + c$   
$y = {\beta_1x} + {\beta_0}$

y = Dependent/Target or the value to be predicted.   
m = gradient or the slope of the curve(how much y changes when x increases by 1).         
x = Independent variable/predictor variable.     
c = y-intercept(the value of y when x is equal to zero where the line crosses y-axis).     

_Example_      
|House size(sqm)(x)|monthly rent(y)|
|------------------|---------------|
|80                |25000          |
|220               |250000         |
|150               |120000         |   


Slope: $\beta_1 = \frac{n\sum XY - (\sum X)(\sum Y)}{n\sum X^2 - (\sum X)^2}$   

Y intercept:  $\beta_0 = \frac{\sum Y - \beta_1\sum X}{n}$  

````python
import pandas as pd

data = {'house_size':[80, 220, 150],
        'monthly_rent':[25000, 250000,120000]}

df = pd.DataFrame(data)

# Multiply X and Y
df["xy"] = df["house_size"] * df["monthly_rent"]

# Total of x
sum_house_size = df["house_size"].sum()

# Total of Y
sum_rent = df["monthly_rent"].sum()

# Square X
df["x_squared"] = df["house_size"]**2

print("="*35, "SLOPE", "="*35)
slope = ((len(df) * (df["xy"].sum())) - ((sum_house_size)*(sum_rent)))/((len(df)*(df["x_squared"].sum()))-(sum_house_size**2))

print(slope)
print("="*35, "Y-INTERCEPT", "="*35)
y_intercept = sum_rent - (1607.142857142857 * sum_house_size)/len(df)
print(y_intercept)
````

__Definition of terms in Linear Regression__

- Intercept:This is the predicted value of Y when x=0.     
- Slope: The slope tells us how much y changes when X increases by one unit.

- Residuals(Errors): Residuals = Actual values  predicted values.

- Least Squares Method: The model choose the regression line tht minimizes the sum squared residuals.  Instead of minimizing(which cancels positive and negative errors):

#### Model Evaluation Metrics.    
After building a linear regression model we need to determine how well it performs. Model evaluation metrics compares the model's predicted values with the actual observed values and quantify the prediction error.    
These metrics help us in answering questions such as:    
- How accurate are the predictions.   
- Which regression model performs better?  
- Does the model explain a large proportion of the variation in target variable?   

##### 1.Mean absolute error.   
The mean Absolute Error(MAE) measure the average magnitude of the prediction errors without considering whether the model overestimates(negative error) or underestimates(positive error) the target variable.  
_Formula_    
$MAE=\frac{1}{n}\sum_{i=1}^{n}|y_i-\hat{y}_i|$

Errors:   
|Actual |Predicted | Residuals|absolute error|
|-------|----------|----------|--------------|
|24     |20        |4         |4             |
|40     |44        |-4        |4             |
|56     |59        |-3        |3             |
|70     |67        |3         |3             |  

$MAE = \frac{4 + 4 + 3 + 3}{3}$ = 3.5   

_Advantages_  
- Very easy to understand.   
- Expressed in the same units as the target variable
- Less sensitive to extreme errors than MSE and RMSE.  
- Suitable when all prediction errors should be treated equally.     

_Disadvantages_   
- Does not penalize large errors more heavily than small ones.   
- Two models with the same MAE may have very different error distributions.

##### 2. Mean Squared Error(MSE)
The mean Square Error(MSE) Calculates the average of the squared prediction errors.    
_Formula_     
$MSE=\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)^2$    

Because the errors are squared:    
- Large errors contribute much more than smaller ones.    
- Negative errors become positive.  
- Models that make a few very large mistakes(errors) recieve a higher penalty.  

|Prediction error| Squared error |
|----------------|---------------|
|2               |4              |
|3               |9              |
|10              |100            |


Notice that the error of 10 contributes 100 to the MSE, making it influential than the smaller errors.    

__Adantages__    
- Penalizes large prediction errors heavily
- Forms the basis of Ordinary Least Squares(OLS) regression.     

__Disadvantages__.      
- Dificult to interpret because units are squared.     
- Highly sensitive to outliers.     

##### 3. Root Mean Squared Error(RMSE)
The Root Mean Squared Error(RMSE) is the square root of the mean squared error.    
The formula is:     
$RMSE = \sqrt{MSE}$    

The RMSE converts the error back to the original units of the target variable.     
_Example_    
Suppose you were predicting the monthly rent in Nairobi.     
MSE = 1000,000       

then the RMSE:    
$RMSE = \sqrt{1000,000} = 1000$  

If your units were in Shillings this means that the model's prediction differ from the actual values by approximately 1000 Shilling on average.      

_Advantages_      
- Easy to interpret because it has the same units as the target variable.     
- Penalizes large prediction errors.    

_Disadvantages_       
- Sensitive to outliers.    
- Can be heavily influenced by a few extreme prediction errors.       


##### 4. R<sup>2</sup> Score(Coefficient of Determination).    
The R2 Score measures how well the regression model explains the variation in the dependent variable.     
_Formula_       
$R^2=1-\frac{SS_{Residual}}{SS_{Total}}$      

- $SS_{Residual}$ = Sum of Squared Residuals(Unexplained Variation).   
$SS_{res}=\sum (y_{i}-\^{y}_{i})^{2}$

- $SS_{Total}$ = Total sum of Squares(Total variation in the target variable).    
$SS_{tot}=\sum (y_{i}-\={y})^{2}$    

_Example_     

|Actual Values|Predicted Values|
|-------------|----------------|
|2            |2.5             |
|3            |2.0             |
|7            |7.5             |   

Calculate the r2 score.

````python
# Step 1: Calculate the mean of the actual values.    
actual_values = [2, 3, 7]

mean_actual_values = sum(actual_values) / len(actual_values)

# Step II: Calculate the total sum of squares and the residual sum squared  
total_sum_of_squares = ((2-mean_actual_values)**2) + ((3 - mean_actual_values)**2) + ((7 - mean_actual_values)**2)

residual_sum_squared = ((2-2.5)**2) + ((3-2.0)**2) + ((7 - 7.5)**2)

# Step III: Calculate the R2 Score
r2_score = 1 - (residual_sum_squared/total_sum_of_squares)

print("Mean of actual values", mean_actual_values)
print("total_sum_of_squares", total_sum_of_squares)
print("Residual Sum Squared",residual_sum_squared)
print("R2 Score", r2_score)
````

A R<sup>2</sup> of 0.89 or 89% means that 89% of the variation in the target variable can be explained by the predictors, while thw remaining 11% is due to factors not captured by the modelor random variation.   

_Interpretations_   
|R<sup>2<\sup>|Interpretation                                                  |
|-------------|----------------------------------------------------------------|
|0.00         |The model explains none of the variation in the target variable | 
|0.40         |The model explains 40% of the variation                         |
|0.75         |The model explains 75% of the variation                         |
|1.0          |Perfect prediction                                              |      

_Advantages_      
- Easy to interpret as a percentage of explained variance.    
- Useful for comparing models built on the same dataset.    
- Provides the overall measure of the model fit.    

_Disadvantages_      
- A high R2 score does not necessarily mean the model is appropriate.    
- Adding more predictors can artificially increase the R2, even if those predictors add little predictive value.    

##### 5. Adjusted R<sup>2</sup>    
The adjusted R<sup>2</sup> improves upon R2 by accounting for the number of predictors in the model.     

Unlike the R2, the adjusted R2 increases only when a new predictor guininely improves the model's performance. If an irrelevant predictor is added, Adjusted R2 may remain unchanged or even decrease.    

_Example_     
Suppose we have two models predicting house prices:     
- __Model A:__ Size, Bedrooms     
- __Model B:__ Size, Bedrooms, House colour.      

House colour is unlikely to influence the house prices. Although adding may slightly increase the R2 value, it provides no meaningful predictive power.     
Adjusted R2 helps identify this by penalizing unnecessary predictors.    

_Advantages_     
- Reduces the risk of overfitting.    
- Better than R2 when comparing models with different number of predictors.    
- Encourages selecting only meaningfulv predictors.     

_Disadvantages._    
- Slightly more difficult to interpret than R2.    


**Note:** A well performing regression model typically has low MAE, Low MSE, low RMSE, and high R<sup>2</sup> and Adjusted R<sup>2
</sup>, while satisfying the assumptions of linear regression.
