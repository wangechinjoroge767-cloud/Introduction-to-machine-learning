# Introduction to Machine Learning.     
## 1. What is machine Learning?    
Machine learning is a branch of AI that allows computers to learn patterns from data and make predictions or decisions without being explicitly programmed for every rule.       
Instead of writing fixed instructions, we give the computer data, and it learns relationships from that data.     

_Example_      
Instead of manually writing rules to determine house rent in Nairobi, we can train a machine learning model using previous housing data.. The model can learn how factors such as estate, property type, number of bedrooms, house size, availabililty of parking, and the distance to cbd influence rent prices.     

|Estate. |Bedrooms|House size(sqm)|Parking|Monthly Rent |
|------  |--------|---------------|-------|-------------|
|Kasarani|2       |80             |No     |25,000       |
|Karen   |4       |220            |Yes    |250,000      |
|Kilimani|3       |150            |Yes    |120,000      |    

A machine learning model can study this data and learn that:     
- Houses i high end estates usually have higher rent.    
- More bedrooms often increase rent.    
- Bigger houses may lead to higher rent.   
- Parking availability can affect price.      

After learning from past data, the model can predict rent for a new house.

## Types of Machine Learning. 
- 1. Supervised learning.  
- 2. Unsupervised learning.  
- 3. Reinforcement learning.    

### 1. Supervised learning. 
This is a type of machine learning where the model learns from labelled data.  
Labelled data s a dataset that already contains the corrrect answer.   

_Example_    
|Estate  |Bedrooms|House size(sqm)|Parking|Monthly Rent |
|------  |--------|---------------|-------|-------------|
|Kasarani|2       |80             |No     |25,000       |
|Karen   |4       |220            |Yes    |250,000      |
|Kilimani|3       |150            |Yes    |120,000      |   

Here:    
__Features:__ Estate, Bedrooms, House Size(sqm), Parking.       
__Label:__ Monthly Rent.      

The model learns the relationship between the features and the label.       
Common supervised learning tasks:     
- __Regression:__ Regression is used when the target variable is numeric. e.g. predicting house rent, predicting salary, predicting sales revenue, predicting student examp score.       

- __Classification:__ Classification is used when the target variable is a category. e.g. predicting if whether a customer will default: Yes or No, predicting whether an email is spam or not spam, predicting whether a student will pass or fail.   

### 2. Unsupervised learning.    
Unsupervised learning is used when the dataset does not have a labelled answer.    
The model tries to discover hidden patterns, groups, or structures in the data by itself.     

_Example_     
Assume we have the housing data with estate, rent, house size, bedrooms, distance to cbd but we do not give the model the target column to predict.   
The model may group houses into clusters such as:   
* Budget houses.   
* Middle income houses 
* Luxury Houses.  

__Common categories in Unsupervised learning__     
- __Clustering:__ Clustering means grouping similar records togethe.    
_Example_     
The model may group houses into clusters such as:   
* Budget houses.   
* Middle income houses 
* Luxury Houses.    

- __Dimensionality reduction:__ Dimensionality reduction reduces many columns into fewer important components while still preserving useful information.     
It is uselful when a dataset has many columns.      

_Example_    
Reducing many housing features into fewer property quality indicators.    

### 3. Reinforcement learning.    
Reinforcement learning is a type of machine learning where a model learns by interacting with enviroment.       
The agent takes actions and recieves penalties or rewards.
